# 구현 3 완전 이진 트리(Heap 자료구조)
-> 우선순위 큐 


enqueue (heapq.push) : O(logn) 
dequeue (heapq.pop) : O(logn) 

```python
# min heap -> 우선순위 큐로 사용하려고(다익스트라)
import heapq
# enqueue
min_heap = [5,3,9,4,1,2,6]

heapq.heapify(min_heap) # H=O(N)
# dequeue
heapq.heappop(min_heap) # H=O(logN)
 
heapq.heappush(min_heap,1) # H=O(logN)
```

```python
# max heap - 1 (heappush x)
import heapq

max_heap = [5,3,9,4,1,2,6]
heapq.heapify_max(max_heap)
value = heapq._heappop_max(max_heap)
```

```python
# max heap -2 (heappush o)
import heapq

max_heap = [5,3,9,4,1,2,6]
max_heap = [i * -1 for i in max_heap]
heapq.heapify(max_heap)
weight = heapq.heappop(max_heap)
value = -1 * weight
```

```python
# max heap -3

import heapq

max_heap = [5,3,9,4,1,2,6]
max_heap = [(-1 * i, i) for i in max_heap]
heapq.heapify(max_heap)
weight, value = heapq.heappop(max_heap)
```
